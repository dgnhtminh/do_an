```python
# ===========================================
# CELL 2: TRICH XUAT CPG BANG JOERN (DA CAI THIEN)
# ===========================================
# Cai thien:
#   - Export voi --repr=all de lay day du vertex properties (CODE, NAME, TYPE...)
#   - Tim file graph bang pattern thay vi lay file lon nhat
#   - Validate GraphSON structure truoc khi luu
#   - Error logging chi tiet vao file

import pandas as pd
import os
import subprocess
import shutil
import sys
import json
import logging

# ==========================================
# CAU HINH
# ==========================================
DRIVE_DIR = "/content/drive/MyDrive/Đồ án"
CSV_PATH = f"{DRIVE_DIR}/verum_dataset.csv"
CODE_COLUMN_NAME = "code"
LABEL_COLUMN_NAME = "label"
BATCH_SIZE = 1000

JOERN_PARSE = "/content/joern/joern-cli/bin/joern-parse"
JOERN_EXPORT = "/content/joern/joern-cli/bin/joern-export"

TRACKER_FILE = f"{DRIVE_DIR}/tien_do.txt"
BATCH_ROOT_DIR = f"{DRIVE_DIR}/cpg_batches"
ERROR_LOG_FILE = f"{DRIVE_DIR}/extraction_errors.log"

# ==========================================
# SETUP LOGGING
# ==========================================
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(message)s',
    handlers=[
        logging.FileHandler(ERROR_LOG_FILE, mode='a', encoding='utf-8'),
        logging.StreamHandler(sys.stdout)
    ]
)
logger = logging.getLogger('cpg_extraction')

# ==========================================
# CHECK JOERN
# ==========================================
if not os.path.exists(JOERN_PARSE):
    logger.error("Khong tim thay joern-parse")
    sys.exit()
if not os.path.exists(JOERN_EXPORT):
    logger.error("Khong tim thay joern-export")
    sys.exit()

os.makedirs(DRIVE_DIR, exist_ok=True)
os.makedirs(BATCH_ROOT_DIR, exist_ok=True)

# ==========================================
# VALIDATE GRAPHSON
# ==========================================
def validate_graphson(json_path):
    """Kiem tra file GraphSON co cau truc hop le."""
    try:
        with open(json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        value = data.get('@value', {})
        has_edges = len(value.get('edges', [])) > 0
        has_vertices = len(value.get('vertices', [])) > 0
        return has_edges or has_vertices
    except (json.JSONDecodeError, KeyError, TypeError):
        return False

# ==========================================
# DOC TIEN DO
# ==========================================
start_row = 0
if os.path.exists(TRACKER_FILE):
    with open(TRACKER_FILE, "r") as f:
        content = f.read().strip()
        if content.isdigit():
            start_row = int(content)

logger.info(f"BAT DAU TU DONG: {start_row}")

# ==========================================
# DOC CSV
# ==========================================
try:
    df = pd.read_csv(CSV_PATH)
except FileNotFoundError:
    logger.error("Khong tim thay CSV.")
    sys.exit()

if LABEL_COLUMN_NAME not in df.columns:
    logger.error("Khong co cot label.")
    sys.exit()

total_rows = len(df)
if start_row >= total_rows:
    logger.info("Da xu ly xong toan bo dataset.")
    sys.exit()

end_row = min(start_row + BATCH_SIZE, total_rows)
logger.info(f"Xu ly den dong: {end_row}")
df_batch = df.iloc[start_row:end_row]

# ==========================================
# TAO BATCH DIR
# ==========================================
BATCH_DIR = os.path.join(BATCH_ROOT_DIR, f"batch_{start_row}_{end_row}")
GRAPH_DIR_DRIVE = os.path.join(BATCH_DIR, "graphs")
TEMP_CODE_DIR = "/content/temp_code"
TEMP_CPG_FILE = "/content/temp.cpg"
LOCAL_GRAPH_DIR = "/content/temp_graph_export"

if os.path.exists(BATCH_DIR):
    logger.warning("Batch da ton tai -> dung.")
    sys.exit()

os.makedirs(BATCH_DIR)
os.makedirs(GRAPH_DIR_DRIVE)

# ==========================================
# PARSE VA EXPORT TUNG FILE
# ==========================================
logger.info("Bat dau xu ly va tao do thi...")
valid_indices = []
error_count = 0

for idx, row in df_batch.iterrows():
    if os.path.exists(TEMP_CODE_DIR): shutil.rmtree(TEMP_CODE_DIR)
    os.makedirs(TEMP_CODE_DIR, exist_ok=True)
    if os.path.exists(LOCAL_GRAPH_DIR): shutil.rmtree(LOCAL_GRAPH_DIR)
    if os.path.exists(TEMP_CPG_FILE): os.remove(TEMP_CPG_FILE)

    code = str(row[CODE_COLUMN_NAME]) if pd.notna(row[CODE_COLUMN_NAME]) else ""
    if not code.strip():
        logger.warning(f"Code rong tai index {idx}, bo qua.")
        continue

    # 1. Ghi file C
    file_path = os.path.join(TEMP_CODE_DIR, "temp.c")
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(code)

    # 2. Parse CPG
    parse_result = subprocess.run(
        [JOERN_PARSE, TEMP_CODE_DIR, "--language", "c", "-o", TEMP_CPG_FILE],
        capture_output=True, text=True, timeout=120
    )
    if not os.path.exists(TEMP_CPG_FILE) or os.path.getsize(TEMP_CPG_FILE) == 0:
        logger.error(f"Parse that bai tai index {idx}: {parse_result.stderr[:200]}")
        error_count += 1
        continue

    # 3. Export GraphSON voi --repr=all de lay day du vertex properties
    export_result = subprocess.run(
        [JOERN_EXPORT, TEMP_CPG_FILE, "--repr", "all", "--format", "graphson", "--out", LOCAL_GRAPH_DIR],
        capture_output=True, text=True, timeout=120
    )

    # 4. Tim file graph, validate truoc khi luu
    if os.path.exists(LOCAL_GRAPH_DIR):
        graph_files = []
        for root_dir, dirs, files in os.walk(LOCAL_GRAPH_DIR):
            for file in files:
                if file.endswith('.json') or file.endswith('.graphson'):
                    graph_files.append(os.path.join(root_dir, file))

        # Uu tien file co ten 'export' hoac 'cpg'
        selected_file = None
        for gf in graph_files:
            basename = os.path.basename(gf).lower()
            if 'export' in basename or 'cpg' in basename:
                selected_file = gf
                break
        if selected_file is None and len(graph_files) > 0:
            graph_files.sort(key=lambda x: os.path.getsize(x), reverse=True)
            selected_file = graph_files[0]

        if selected_file is not None and validate_graphson(selected_file):
            dest_path = os.path.join(GRAPH_DIR_DRIVE, f"graph_{idx}.json")
            shutil.copy(selected_file, dest_path)
            valid_indices.append(idx)
            if len(valid_indices) % 100 == 0:
                logger.info(f"Da xu ly {len(valid_indices)} graphs thanh cong...")
        else:
            logger.error(f"GraphSON khong hop le tai index {idx}")
            error_count += 1
    else:
        logger.error(f"Khong tao duoc thu muc export tai index {idx}")
        error_count += 1

logger.info(f"Xu ly xong. Thanh cong: {len(valid_indices)}/{len(df_batch)}, Loi: {error_count}")

# ==========================================
# LUU METADATA + UPDATE TRACKER
# ==========================================
if len(valid_indices) > 0:
    metadata = df_batch.loc[valid_indices, [LABEL_COLUMN_NAME]].copy()
    metadata = metadata.reset_index()
    metadata.rename(columns={"index": "original_index"}, inplace=True)
    metadata["graph_file"] = metadata["original_index"].apply(lambda x: f"graph_{x}.json")
    metadata = metadata[["original_index", "graph_file", LABEL_COLUMN_NAME]]
    metadata.to_csv(os.path.join(BATCH_DIR, "metadata.csv"), index=False)
    logger.info("Da luu metadata.csv")

with open(TRACKER_FILE, "w") as f:
    f.write(str(end_row))

if os.path.exists(TEMP_CODE_DIR): shutil.rmtree(TEMP_CODE_DIR)
if os.path.exists(TEMP_CPG_FILE): os.remove(TEMP_CPG_FILE)
if os.path.exists(LOCAL_GRAPH_DIR): shutil.rmtree(LOCAL_GRAPH_DIR)
logger.info(f"HOAN THANH BATCH: {BATCH_DIR}")
```

```python
# ===========================================
# CELL 4: TIEN XU LY DO THI (DA TOI UU TOC DO)
# ===========================================
# Toi uu:
#   - Copy data tu Drive sang local disk TRUOC khi xu ly (nhanh hon 10-100x)
#   - Pre-compile regex (tranh compile lai moi lan goi)
#   - Progress tracking voi ETA
#   - Parse vertices + edges, rich node features

import os
import json
import re
import time
import torch
import pandas as pd
import shutil
import logging
from collections import Counter
from torch_geometric.data import Data

logger = logging.getLogger('preprocessing')
logging.basicConfig(level=logging.INFO, format='%(asctime)s [%(levelname)s] %(message)s')

# ==========================================
# CAU HINH
# ==========================================
DRIVE_DIR = "/content/drive/MyDrive/Đồ án"
BATCH_ROOT_DIR_DRIVE = f"{DRIVE_DIR}/cpg_batches"   # Tren Drive (CHAM)
BATCH_ROOT_DIR_LOCAL = "/content/cpg_batches_local"  # Tren local SSD (NHANH)
LABEL_COLUMN_NAME = "label"
LOCAL_PROCESSED_DIR = "/content/processed_graphs_local"
DRIVE_ZIP_PATH = f"{DRIVE_DIR}/dataset_cpg_v2.zip"
VOCAB_PATH = f"{DRIVE_DIR}/vocabularies_v2.json"
MAX_TEST_FILES = None  # None = toan bo
MAX_CODE_TOKENS = 32   # Giam tu 64 xuong 32 de nhanh hon, it anh huong quality
CODE_VOCAB_MAX_SIZE = 10000

# ==========================================
# BUOC 0: COPY DATA TU DRIVE SANG LOCAL (FIX CHINH)
# ==========================================
# Google Drive I/O rat cham (~100-500ms/file).
# Copy 1 lan sang local SSD roi xu ly tren do -> nhanh hon 10-100x.
print("="*60)
print("BUOC 0: Copy data tu Drive sang local disk...")
print("="*60)

if os.path.exists(BATCH_ROOT_DIR_LOCAL):
    shutil.rmtree(BATCH_ROOT_DIR_LOCAL)

t0 = time.time()
shutil.copytree(BATCH_ROOT_DIR_DRIVE, BATCH_ROOT_DIR_LOCAL)
copy_time = time.time() - t0
print(f"Copy xong trong {copy_time:.1f}s")

# Don dep output dir
if os.path.exists(LOCAL_PROCESSED_DIR):
    shutil.rmtree(LOCAL_PROCESSED_DIR)
os.makedirs(LOCAL_PROCESSED_DIR, exist_ok=True)

# ==========================================
# VOCABULARY
# ==========================================
node_type_vocab = {}
edge_type_vocab = {}
code_token_vocab = {'<PAD>': 0, '<UNK>': 1}
name_vocab = {'<UNK>': 0}

def get_vocab_id(vocab_dict, key, max_size=None):
    if key not in vocab_dict:
        if max_size is not None and len(vocab_dict) >= max_size:
            return vocab_dict.get('<UNK>', 0)
        vocab_dict[key] = len(vocab_dict)
    return vocab_dict[key]

# ==========================================
# CODE TOKENIZER (PRE-COMPILED REGEX)
# ==========================================
CODE_TOKEN_PATTERN = re.compile(r'[a-zA-Z_]\w*|[0-9]+|[^\s\w]')

def tokenize_code(code_str, max_tokens=MAX_CODE_TOKENS):
    if not code_str or not isinstance(code_str, str):
        return [0] * max_tokens
    tokens = CODE_TOKEN_PATTERN.findall(code_str)
    tokens = [t.lower() for t in tokens[:max_tokens]]
    token_ids = [get_vocab_id(code_token_vocab, t, max_size=CODE_VOCAB_MAX_SIZE) for t in tokens]
    pad_len = max_tokens - len(token_ids)
    if pad_len > 0:
        token_ids += [0] * pad_len
    return token_ids[:max_tokens]

# ==========================================
# EXTRACT GRAPHSON PROPERTY
# ==========================================
def get_property_value(properties, key, default=''):
    prop = properties.get(key, [])
    if isinstance(prop, list) and len(prop) > 0:
        val = prop[0]
        if isinstance(val, dict):
            v = val.get('@value', val.get('value', default))
            if isinstance(v, dict):
                return str(v.get('@value', default))
            return str(v)
        return str(val)
    elif isinstance(prop, dict):
        return str(prop.get('@value', prop.get('value', default)))
    return str(default)

# ==========================================
# CHUYEN GRAPHSON SANG TENSOR
# ==========================================
def process_graphson_to_tensor(json_path, label):
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    value = data.get('@value', data)
    vertices_data = value.get('vertices', [])
    edges_data = value.get('edges', [])

    if not edges_data:
        return None

    # --- Parse VERTICES ---
    node_id_map = {}
    node_info = {}
    current_node_idx = 0

    for vertex in vertices_data:
        v_id = vertex.get('id', vertex.get('@value', {}).get('id'))
        if isinstance(v_id, dict):
            v_id = v_id.get('@value', v_id)

        v_label = vertex.get('label', 'UNKNOWN')
        properties = vertex.get('properties', {})

        code_str = get_property_value(properties, 'CODE', '')
        name_str = get_property_value(properties, 'NAME', '')
        order_val = get_property_value(properties, 'ORDER', '0')
        line_num = get_property_value(properties, 'LINE_NUMBER', '0')

        try:
            order_int = int(order_val) if order_val.isdigit() else 0
        except (ValueError, AttributeError):
            order_int = 0
        try:
            line_int = int(line_num) if line_num.isdigit() else 0
        except (ValueError, AttributeError):
            line_int = 0

        if v_id not in node_id_map:
            node_id_map[v_id] = current_node_idx
            node_info[current_node_idx] = {
                'type': get_vocab_id(node_type_vocab, v_label),
                'code_tokens': tokenize_code(code_str),
                'name': get_vocab_id(name_vocab, name_str if name_str else '<UNK>'),
                'order': min(order_int, 999),
                'line_num': min(line_int, 9999),
            }
            current_node_idx += 1

    # --- Parse EDGES ---
    source_nodes, target_nodes, edge_attr_list = [], [], []

    for edge in edges_data:
        out_v = edge.get('outV')
        in_v = edge.get('inV')
        out_id = out_v.get('@value') if isinstance(out_v, dict) else out_v
        in_id = in_v.get('@value') if isinstance(in_v, dict) else in_v
        edge_label = edge.get('label', 'UNKNOWN')

        if out_id not in node_id_map:
            out_label = edge.get('outVLabel', 'UNKNOWN')
            node_id_map[out_id] = current_node_idx
            node_info[current_node_idx] = {
                'type': get_vocab_id(node_type_vocab, out_label),
                'code_tokens': [0] * MAX_CODE_TOKENS,
                'name': 0, 'order': 0, 'line_num': 0,
            }
            current_node_idx += 1

        if in_id not in node_id_map:
            in_label = edge.get('inVLabel', 'UNKNOWN')
            node_id_map[in_id] = current_node_idx
            node_info[current_node_idx] = {
                'type': get_vocab_id(node_type_vocab, in_label),
                'code_tokens': [0] * MAX_CODE_TOKENS,
                'name': 0, 'order': 0, 'line_num': 0,
            }
            current_node_idx += 1

        source_nodes.append(node_id_map[out_id])
        target_nodes.append(node_id_map[in_id])
        edge_attr_list.append(get_vocab_id(edge_type_vocab, edge_label))

    if current_node_idx == 0:
        return None

    # --- Tao TENSORS ---
    edge_index = torch.tensor([source_nodes, target_nodes], dtype=torch.long)
    edge_attr = torch.tensor(edge_attr_list, dtype=torch.long).view(-1, 1)
    y = torch.tensor([label], dtype=torch.float)

    node_scalar_features = torch.tensor(
        [[node_info[i]['type'], node_info[i]['name'],
          node_info[i]['order'], node_info[i]['line_num']]
         for i in range(current_node_idx)],
        dtype=torch.long
    )

    code_tokens = torch.tensor(
        [node_info[i]['code_tokens'] for i in range(current_node_idx)],
        dtype=torch.long
    )

    return Data(
        x=node_scalar_features,
        code_tokens=code_tokens,
        edge_index=edge_index,
        edge_attr=edge_attr,
        y=y
    )

# ==========================================
# DEM TONG SO FILE CAN XU LY
# ==========================================
batch_dirs = [os.path.join(BATCH_ROOT_DIR_LOCAL, d)
              for d in os.listdir(BATCH_ROOT_DIR_LOCAL) if d.startswith('batch_')]
batch_dirs.sort(key=lambda x: int(os.path.basename(x).split('_')[1]))

total_files = 0
for bd in batch_dirs:
    mp = os.path.join(bd, 'metadata.csv')
    if os.path.exists(mp):
        total_files += len(pd.read_csv(mp))
print(f"\nTong so file can xu ly: {total_files}")

# ==========================================
# CHAY XU LY (TU LOCAL DISK - NHANH)
# ==========================================
print("\n" + "="*60)
print("BAT DAU XU LY (tu local disk)")
print("="*60)

total_processed = 0
total_errors = 0
start_time = time.time()

for batch_dir in batch_dirs:
    if MAX_TEST_FILES is not None and total_processed >= MAX_TEST_FILES:
        break

    metadata_path = os.path.join(batch_dir, 'metadata.csv')
    graphs_dir = os.path.join(batch_dir, 'graphs')
    if not os.path.exists(metadata_path) or not os.path.exists(graphs_dir):
        continue

    df_meta = pd.read_csv(metadata_path)
    df_meta = df_meta.sort_values('original_index').reset_index(drop=True)
    batch_name = os.path.basename(batch_dir)

    for _, row in df_meta.iterrows():
        if MAX_TEST_FILES is not None and total_processed >= MAX_TEST_FILES:
            break
        orig_idx = row['original_index']
        graph_file = row['graph_file']
        label = row[LABEL_COLUMN_NAME]
        json_path = os.path.join(graphs_dir, graph_file)

        if os.path.exists(json_path):
            try:
                graph_data = process_graphson_to_tensor(json_path, label)
                if graph_data is not None:
                    save_path = os.path.join(LOCAL_PROCESSED_DIR, f'data_{orig_idx}.pt')
                    torch.save(graph_data, save_path)
                    total_processed += 1
                else:
                    total_errors += 1
            except Exception as e:
                logger.error(f'Loi index {orig_idx}: {type(e).__name__}: {e}')
                total_errors += 1

        # Progress tracking moi 100 files
        done = total_processed + total_errors
        if done > 0 and done % 100 == 0:
            elapsed = time.time() - start_time
            speed = done / elapsed
            remaining = (total_files - done) / speed if speed > 0 else 0
            print(f'  [{batch_name}] {done}/{total_files} '
                  f'({done/total_files*100:.1f}%) | '
                  f'{speed:.1f} files/s | '
                  f'ETA: {remaining/60:.1f} phut')

elapsed_total = time.time() - start_time
print(f'\nXu ly xong trong {elapsed_total:.1f}s ({elapsed_total/60:.1f} phut)')
print(f'Thanh cong: {total_processed}, Loi: {total_errors}')

# ==========================================
# LUU VOCABULARY + NEN ZIP (len Drive)
# ==========================================
print("\nLuu vocabulary va nen zip...")

with open(VOCAB_PATH, 'w', encoding='utf-8') as f:
    json.dump({
        'node_types': node_type_vocab,
        'edge_types': edge_type_vocab,
        'code_tokens': code_token_vocab,
        'names': name_vocab,
        'config': {
            'max_code_tokens': MAX_CODE_TOKENS,
            'code_vocab_max_size': CODE_VOCAB_MAX_SIZE,
        }
    }, f, ensure_ascii=False, indent=2)

zip_base = '/content/dataset_cpg_v2'
shutil.make_archive(zip_base, 'zip', LOCAL_PROCESSED_DIR)
shutil.copy(zip_base + '.zip', DRIVE_ZIP_PATH)

# Don dep local copy
shutil.rmtree(BATCH_ROOT_DIR_LOCAL, ignore_errors=True)

print('=' * 60)
print(f'HOAN TAT!')
print(f'Dataset: {DRIVE_ZIP_PATH}')
print(f'Vocabulary: {VOCAB_PATH}')
print(f'Node types: {len(node_type_vocab)}, Edge types: {len(edge_type_vocab)}')
print(f'Code tokens: {len(code_token_vocab)}, Names: {len(name_vocab)}')
print(f'Total graphs: {total_processed}')
print('=' * 60)
```

```python
!pip install torch torch_geometric -q

import torch
import torch.nn as nn # Fixed: Added this import

# ===========================================
# CELL 8: MODEL + LOSS + HELPERS (FIX IMBALANCE)
# ===========================================
# Cai thien xu ly imbalance:
#   1. Focal Loss thay BCEWithLogitsLoss (giam anh huong cua easy samples)
#   2. Helper functions track Vuln F1 rieng (metric quan trong nhat)
#   3. Threshold tuning sau training

# ==========================================
# FOCAL LOSS (MOI)
# ==========================================
class FocalLoss(nn.Module):
    """Focal Loss cho imbalanced binary classification.

    Focal Loss giam trong so cua 'easy' samples (Safe predict dung de)
    va tap trung vao 'hard' samples (Vulnerable kho phan biet).

    alpha: trong so cho class positive (Vulnerable). alpha > 0.5 tang trong so Vuln.
    gamma: he so focusing. gamma > 0 giam loss cua easy samples.
           gamma=0 -> tuong duong BCE. gamma=2 la gia tri pho bien.
    """
    def __init__(self, alpha=0.75, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, logits, targets):
        probs = torch.sigmoid(logits)
        # Binary cross entropy per sample
        bce = F.binary_cross_entropy_with_logits(logits, targets, reduction='none')
        # p_t = prob of correct class
        p_t = probs * targets + (1 - probs) * (1 - targets)
        # alpha weighting: alpha cho positive, (1-alpha) cho negative
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        # Focal term: giam loss cua easy samples
        focal_weight = alpha_t * (1 - p_t) ** self.gamma
        loss = focal_weight * bce
        return loss.mean()

# ==========================================
# MODEL DEFINITION (giu nguyen)
# ==========================================
class CodeTokenEncoder(nn.Module):
    """Encode code token sequences bang multi-kernel 1D CNN."""
    def __init__(self, vocab_size, emb_dim=32, out_dim=64, kernel_sizes=[3, 5, 7]):
        super().__init__()
        self.token_emb = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        per_kernel = out_dim // len(kernel_sizes)
        self.convs = nn.ModuleList([
            nn.Conv1d(emb_dim, per_kernel, k, padding=k // 2)
            for k in kernel_sizes
        ])
        self.out_dim = per_kernel * len(kernel_sizes)

    def forward(self, token_ids):
        x = self.token_emb(token_ids)
        x = x.transpose(1, 2)
        conv_outs = []
        for conv in self.convs:
            c = F.relu(conv(x))
            c = c.max(dim=2)[0]
            conv_outs.append(c)
        return torch.cat(conv_outs, dim=1)


class AttentionReadout(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.att_net = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, x, batch):
        att_scores = self.att_net(x)
        att_weights = pyg_softmax(att_scores, batch, dim=0)
        weighted_x = x * att_weights
        return global_add_pool(weighted_x, batch)


class VulnRGCN(nn.Module):
    def __init__(self, num_node_types, num_edge_types, hidden_dim=128,
                 num_layers=3, dropout=0.3,
                 num_code_tokens=10000, code_emb_dim=32,
                 num_names=5000, max_code_tokens=64):
        super().__init__()
        self.node_type_emb = nn.Embedding(num_node_types, hidden_dim // 4)
        self.name_emb = nn.Embedding(num_names, hidden_dim // 8)
        self.code_encoder = CodeTokenEncoder(
            vocab_size=num_code_tokens, emb_dim=code_emb_dim,
            out_dim=hidden_dim // 2, kernel_sizes=[3, 5, 7]
        )
        self.pos_proj = nn.Linear(2, hidden_dim // 8)

        feature_dim = (hidden_dim // 4 + self.code_encoder.out_dim
                       + hidden_dim // 8 + hidden_dim // 8)
        self.feature_proj = nn.Sequential(
            nn.Linear(feature_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout)
        )

        self.convs = nn.ModuleList()
        self.bns = nn.ModuleList()
        for _ in range(num_layers):
            self.convs.append(RGCNConv(hidden_dim, hidden_dim, num_relations=num_edge_types))
            self.bns.append(nn.BatchNorm1d(hidden_dim))
        self.dropout = dropout

        self.attention_readout = AttentionReadout(hidden_dim)

        self.classifier = nn.Sequential(
            nn.Linear(hidden_dim * 3, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(dropout / 2),
            nn.Linear(hidden_dim // 2, 1)
        )

    def forward(self, data):
        x_scalar = data.x
        type_emb = self.node_type_emb(x_scalar[:, 0])
        if hasattr(data, 'code_tokens') and data.code_tokens is not None:
            code_emb = self.code_encoder(data.code_tokens)
        else:
            code_emb = torch.zeros(x_scalar.shape[0], self.code_encoder.out_dim,
                                  device=x_scalar.device)
        name_emb = self.name_emb(x_scalar[:, 1])
        pos_emb = self.pos_proj(x_scalar[:, 2:4].float())

        x = torch.cat([type_emb, code_emb, name_emb, pos_emb], dim=1)
        x = self.feature_proj(x)

        edge_index = data.edge_index
        edge_type = data.edge_attr.squeeze(-1).long()
        batch = data.batch

        for conv, bn in zip(self.convs, self.bns):
            x_res = x
            x = conv(x, edge_index, edge_type)
            x = bn(x)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            x = x + x_res

        x_att = self.attention_readout(x, batch)
        x_mean = global_mean_pool(x, batch)
        x_max = global_max_pool(x, batch)
        x_graph = torch.cat([x_att, x_mean, x_max], dim=1)
        return self.classifier(x_graph).squeeze(-1)


# ==========================================
# HELPER FUNCTIONS (CAI THIEN: track Vuln F1 rieng)
# ==========================================
def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    all_preds, all_labels, all_probs = [], [], []
    for batch in loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        out = model(batch)
        loss = criterion(out, batch.y.squeeze())
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item() * batch.num_graphs
        probs = torch.sigmoid(out)
        preds = (probs > 0.5).long()
        all_preds.extend(preds.cpu().tolist())
        all_probs.extend(probs.cpu().tolist())
        all_labels.extend(batch.y.squeeze().cpu().long().tolist())
    avg_loss = total_loss / len(loader.dataset)
    acc = accuracy_score(all_labels, all_preds)
    f1_w = f1_score(all_labels, all_preds, average='weighted', zero_division=0)
    # Vuln F1 rieng (metric quan trong nhat)
    f1_vuln = f1_score(all_labels, all_preds, pos_label=1, zero_division=0)
    return avg_loss, acc, f1_w, f1_vuln

@torch.no_grad()
def eval_epoch(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    all_preds, all_labels, all_probs = [], [], []
    for batch in loader:
        batch = batch.to(device)
        out = model(batch)
        loss = criterion(out, batch.y.squeeze())
        total_loss += loss.item() * batch.num_graphs
        probs = torch.sigmoid(out)
        preds = (probs > 0.5).long()
        all_preds.extend(preds.cpu().tolist())
        all_probs.extend(probs.cpu().tolist())
        all_labels.extend(batch.y.squeeze().cpu().long().tolist())
    avg_loss = total_loss / len(loader.dataset)
    acc = accuracy_score(all_labels, all_preds)
    f1_w = f1_score(all_labels, all_preds, average='weighted', zero_division=0)
    f1_vuln = f1_score(all_labels, all_preds, pos_label=1, zero_division=0)
    return avg_loss, acc, f1_w, f1_vuln, all_preds, all_labels, all_probs


def find_best_threshold(labels, probs, metric='f1'):
    """Tim threshold toi uu cho Vuln F1 tren validation set."""
    best_thresh, best_score = 0.5, 0.0
    for thresh in np.arange(0.1, 0.9, 0.02):
        preds = [1 if p > thresh else 0 for p in probs]
        score = f1_score(labels, preds, pos_label=1, zero_division=0)
        if score > best_score:
            best_score = score
            best_thresh = thresh
    return best_thresh, best_score


print('Model, FocalLoss, helpers da duoc dinh nghia.')
```

```python
# ===========================================
# CELL 9: SANITY CHECK (FIX IMBALANCE)
# ===========================================
# Thay doi: Dung Focal Loss, track Vuln F1 rieng

safe_data = [g for g in all_data if int(g.y.item()) == 0]
vuln_data = [g for g in all_data if int(g.y.item()) == 1]
n_per_class = min(SANITY_SIZE // 2, len(safe_data), len(vuln_data))

sanity_data = random.sample(safe_data, n_per_class) + random.sample(vuln_data, n_per_class)
random.shuffle(sanity_data)
print(f'Sanity check voi {len(sanity_data)} samples ({n_per_class} moi class)')

sanity_labels = [int(g.y.item()) for g in sanity_data]
train_idx, val_idx = train_test_split(range(len(sanity_data)), test_size=0.2,
                                       stratify=sanity_labels, random_state=SEED)
train_subset = [sanity_data[i] for i in train_idx]
val_subset = [sanity_data[i] for i in val_idx]

train_loader = DataLoader(train_subset, batch_size=SANITY_BATCH, shuffle=True)
val_loader = DataLoader(val_subset, batch_size=SANITY_BATCH)

model = VulnRGCN(
    num_node_types=num_node_types, num_edge_types=num_edge_types,
    hidden_dim=64, num_layers=2, dropout=0.1,
    num_code_tokens=max(num_code_tokens, 1), code_emb_dim=16,
    num_names=max(num_names, 1), max_code_tokens=max_code_tokens
).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=SANITY_LR, weight_decay=1e-5)
criterion = FocalLoss(alpha=0.75, gamma=2.0)  # Focal Loss thay vi BCE

print(f'\nModel: Enhanced RGCN + Focal Loss (alpha=0.75, gamma=2.0)')
print(f'Parameters: {sum(p.numel() for p in model.parameters()):,}')
print(f'Training...\n')

train_losses, val_losses = [], []
train_f1s_vuln, val_f1s_vuln = [], []

for epoch in range(1, SANITY_EPOCHS + 1):
    t_loss, t_acc, t_f1w, t_f1v = train_epoch(model, train_loader, optimizer, criterion, device)
    v_loss, v_acc, v_f1w, v_f1v, _, _, _ = eval_epoch(model, val_loader, criterion, device)
    train_losses.append(t_loss); val_losses.append(v_loss)
    train_f1s_vuln.append(t_f1v); val_f1s_vuln.append(v_f1v)

    if epoch % 5 == 0 or epoch == 1:
        print(f'Epoch {epoch:3d} | Train L:{t_loss:.4f} A:{t_acc:.4f} VulnF1:{t_f1v:.4f} | '
              f'Val L:{v_loss:.4f} A:{v_acc:.4f} VulnF1:{v_f1v:.4f}')

print('\n' + '=' * 60)
print('KET QUA SANITY CHECK')
print('=' * 60)
loss_ok = train_losses[-1] < train_losses[0]
print(f'Loss giam: {train_losses[0]:.4f} -> {train_losses[-1]:.4f} {"[OK]" if loss_ok else "[FAIL]"}')
print(f'Val Vuln F1: {val_f1s_vuln[-1]:.4f} {"[OK]" if val_f1s_vuln[-1] > 0.4 else "[CANH BAO]"}')
print(f'Best Val Vuln F1: {max(val_f1s_vuln):.4f}')

_, _, _, _, vp, vl, _ = eval_epoch(model, val_loader, criterion, device)
print(f'\nClassification Report (Val):')
print(classification_report(vl, vp, target_names=['Safe', 'Vulnerable'], zero_division=0))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(train_losses, label='Train'); axes[0].plot(val_losses, label='Val')
axes[0].set_title('Loss'); axes[0].legend()
axes[1].plot(train_f1s_vuln, label='Train Vuln F1'); axes[1].plot(val_f1s_vuln, label='Val Vuln F1')
axes[1].set_title('Vulnerable F1-Score (metric chinh)'); axes[1].legend()
axes[1].axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)
plt.tight_layout(); plt.show()

if loss_ok and max(val_f1s_vuln) > 0.3:
    print('\n>>> SANITY CHECK PASSED')
else:
    print('\n>>> SANITY CHECK CO VAN DE')
```

```python
# ===========================================
# CELL 10: FULL TRAINING (FIX IMBALANCE)
# ===========================================
# Fix imbalance:
#   1. Focal Loss (alpha=0.75, gamma=2.0) thay vi BCEWithLogitsLoss
#   2. Class-balanced sampler: oversample Vuln trong moi batch
#   3. Early stopping dua tren Vuln F1 (thay vi weighted F1)
#   4. Threshold tuning sau training
#   5. Epoch nhieu hon (100) vi Focal Loss hoi tu cham hon

from torch.utils.data import WeightedRandomSampler

# --- Split data ---
all_labels = [int(g.y.item()) for g in all_data]
train_idx, temp_idx = train_test_split(range(len(all_data)), test_size=0.3,
                                        stratify=all_labels, random_state=SEED)
temp_labels = [all_labels[i] for i in temp_idx]
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5,
                                      stratify=temp_labels, random_state=SEED)

train_set = [all_data[i] for i in train_idx]
val_set = [all_data[i] for i in val_idx]
test_set = [all_data[i] for i in test_idx]
print(f'Train: {len(train_set)}, Val: {len(val_set)}, Test: {len(test_set)}')

# --- Class-balanced sampler (MOI) ---
# Cho moi sample 1 weight: Vuln duoc sample nhieu hon Safe
train_labels_list = [int(train_set[i].y.item()) for i in range(len(train_set))]
n0 = train_labels_list.count(0)
n1 = train_labels_list.count(1)
class_weights = {0: 1.0 / n0, 1: 1.0 / n1}
sample_weights = [class_weights[l] for l in train_labels_list]
sampler = WeightedRandomSampler(sample_weights, num_samples=len(train_set), replacement=True)

print(f'Class balance: Safe={n0}, Vuln={n1}, ratio={n0/max(n1,1):.1f}')
print(f'Sampler: WeightedRandomSampler (Vuln duoc oversample {n0/max(n1,1):.1f}x)')

# DataLoader voi sampler (khong dung shuffle khi co sampler)
train_loader = DataLoader(train_set, batch_size=FULL_BATCH, sampler=sampler)
val_loader = DataLoader(val_set, batch_size=FULL_BATCH)
test_loader = DataLoader(test_set, batch_size=FULL_BATCH)

# --- Model ---
model_full = VulnRGCN(
    num_node_types=num_node_types, num_edge_types=num_edge_types,
    hidden_dim=HIDDEN_DIM, num_layers=NUM_LAYERS, dropout=DROPOUT,
    num_code_tokens=max(num_code_tokens, 1), code_emb_dim=CODE_EMB_DIM,
    num_names=max(num_names, 1), max_code_tokens=max_code_tokens
).to(device)

optimizer = torch.optim.Adam(model_full.parameters(), lr=FULL_LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5)

# Focal Loss thay vi BCEWithLogitsLoss
criterion = FocalLoss(alpha=0.75, gamma=2.0)

FULL_EPOCHS_NEW = 100  # Tang epoch vi Focal Loss hoi tu cham hon
PATIENCE_NEW = 15

print(f'Model: Enhanced RGCN - {num_edge_types} relations')
print(f'Loss: Focal Loss (alpha=0.75, gamma=2.0)')
print(f'Parameters: {sum(p.numel() for p in model_full.parameters()):,}')
print(f'Epochs: {FULL_EPOCHS_NEW}, Patience: {PATIENCE_NEW}')
print(f'\nTraining...\n')

best_val_vuln_f1 = 0.0
patience_counter = 0
best_epoch = 0
history = {'train_loss': [], 'val_loss': [],
           'train_acc': [], 'val_acc': [],
           'train_f1w': [], 'val_f1w': [],
           'train_f1v': [], 'val_f1v': [],  # Vuln F1
           'lr': []}

for epoch in range(1, FULL_EPOCHS_NEW + 1):
    t_loss, t_acc, t_f1w, t_f1v = train_epoch(model_full, train_loader, optimizer, criterion, device)
    v_loss, v_acc, v_f1w, v_f1v, _, _, _ = eval_epoch(model_full, val_loader, criterion, device)
    lr = optimizer.param_groups[0]['lr']
    scheduler.step(v_f1v)  # Schedule dua tren VULN F1

    history['train_loss'].append(t_loss); history['val_loss'].append(v_loss)
    history['train_acc'].append(t_acc); history['val_acc'].append(v_acc)
    history['train_f1w'].append(t_f1w); history['val_f1w'].append(v_f1w)
    history['train_f1v'].append(t_f1v); history['val_f1v'].append(v_f1v)
    history['lr'].append(lr)

    # Early stopping dua tren VULN F1
    if v_f1v > best_val_vuln_f1:
        best_val_vuln_f1 = v_f1v
        patience_counter = 0
        best_state = {k: v.cpu().clone() for k, v in model_full.state_dict().items()}
        best_epoch = epoch
    else:
        patience_counter += 1

    if epoch % 5 == 0 or epoch == 1:
        print(f'Epoch {epoch:3d} | Train L:{t_loss:.4f} VulnF1:{t_f1v:.4f} | '
              f'Val L:{v_loss:.4f} VulnF1:{v_f1v:.4f} Acc:{v_acc:.4f} | LR:{lr:.6f}')

    if patience_counter >= PATIENCE_NEW:
        print(f'\nEarly stopping at epoch {epoch}! Best: {best_epoch} (VulnF1={best_val_vuln_f1:.4f})')
        break

model_full.load_state_dict(best_state)
model_full = model_full.to(device)

# ==========================================
# THRESHOLD TUNING (MOI)
# ==========================================
print('\n' + '=' * 60)
print('THRESHOLD TUNING')
print('=' * 60)

# Tim threshold toi uu tren validation set
_, _, _, _, val_preds_default, val_labels_all, val_probs_all = eval_epoch(
    model_full, val_loader, criterion, device)
best_thresh, best_thresh_f1 = find_best_threshold(val_labels_all, val_probs_all)

print(f'Default threshold (0.50) -> Vuln F1: {f1_score(val_labels_all, val_preds_default, pos_label=1, zero_division=0):.4f}')
print(f'Optimal threshold ({best_thresh:.2f}) -> Vuln F1: {best_thresh_f1:.4f}')

# ==========================================
# DANH GIA TREN TEST SET (2 threshold)
# ==========================================
print('\n' + '=' * 60)
print('KET QUA TREN TEST SET')
print('=' * 60)

test_loss, test_acc, test_f1w, test_f1v, test_preds, test_labels, test_probs = eval_epoch(
    model_full, test_loader, criterion, device)

# Apply optimal threshold
test_preds_opt = [1 if p > best_thresh else 0 for p in test_probs]
test_f1v_opt = f1_score(test_labels, test_preds_opt, pos_label=1, zero_division=0)
test_acc_opt = accuracy_score(test_labels, test_preds_opt)

print(f'\n--- Threshold = 0.50 (default) ---')
print(f'Test Acc: {test_acc:.4f}, Vuln F1: {test_f1v:.4f}')
print(classification_report(test_labels, test_preds, target_names=['Safe', 'Vulnerable'], zero_division=0))

print(f'--- Threshold = {best_thresh:.2f} (tuned) ---')
print(f'Test Acc: {test_acc_opt:.4f}, Vuln F1: {test_f1v_opt:.4f}')
print(classification_report(test_labels, test_preds_opt, target_names=['Safe', 'Vulnerable'], zero_division=0))

# ROC-AUC, PR-AUC
try:
    roc_auc = roc_auc_score(test_labels, test_probs)
    pr_auc = average_precision_score(test_labels, test_probs)
    print(f'ROC-AUC: {roc_auc:.4f}')
    print(f'PR-AUC: {pr_auc:.4f}')
except ValueError:
    roc_auc, pr_auc = 0.0, 0.0

# --- VISUALIZATION (6 plots) ---
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 1. CM default threshold
cm = confusion_matrix(test_labels, test_preds)
axes[0,0].imshow(cm, cmap='Blues')
axes[0,0].set_xticks([0,1]); axes[0,0].set_yticks([0,1])
axes[0,0].set_xticklabels(['Safe','Vuln']); axes[0,0].set_yticklabels(['Safe','Vuln'])
axes[0,0].set_xlabel('Predicted'); axes[0,0].set_ylabel('Actual')
axes[0,0].set_title(f'CM (threshold=0.50)')
for i in range(2):
    for j in range(2):
        axes[0,0].text(j, i, str(cm[i][j]), ha='center', va='center', fontsize=14)

# 2. CM optimal threshold
cm_opt = confusion_matrix(test_labels, test_preds_opt)
axes[0,1].imshow(cm_opt, cmap='Greens')
axes[0,1].set_xticks([0,1]); axes[0,1].set_yticks([0,1])
axes[0,1].set_xticklabels(['Safe','Vuln']); axes[0,1].set_yticklabels(['Safe','Vuln'])
axes[0,1].set_xlabel('Predicted'); axes[0,1].set_ylabel('Actual')
axes[0,1].set_title(f'CM (threshold={best_thresh:.2f} tuned)')
for i in range(2):
    for j in range(2):
        axes[0,1].text(j, i, str(cm_opt[i][j]), ha='center', va='center', fontsize=14)

# 3. Loss
axes[0,2].plot(history['train_loss'], label='Train')
axes[0,2].plot(history['val_loss'], label='Val')
axes[0,2].set_title('Loss (Focal)'); axes[0,2].legend()

# 4. Vuln F1 (metric chinh)
axes[1,0].plot(history['train_f1v'], label='Train Vuln F1', color='red')
axes[1,0].plot(history['val_f1v'], label='Val Vuln F1', color='darkred')
axes[1,0].set_title('Vulnerable F1-Score (METRIC CHINH)'); axes[1,0].legend()
axes[1,0].axhline(y=0.33, color='gray', linestyle='--', alpha=0.5, label='Baseline (truoc fix)')
axes[1,0].legend()

# 5. ROC Curve
try:
    fpr, tpr, _ = roc_curve(test_labels, test_probs)
    axes[1,1].plot(fpr, tpr, label=f'AUC={roc_auc:.4f}')
    axes[1,1].plot([0,1],[0,1],'k--',alpha=0.5)
    axes[1,1].set_title('ROC Curve'); axes[1,1].legend()
    axes[1,1].set_xlabel('FPR'); axes[1,1].set_ylabel('TPR')
except: axes[1,1].text(0.5,0.5,'N/A',ha='center')

# 6. PR Curve
try:
    prec_v, rec_v, thresholds_pr = precision_recall_curve(test_labels, test_probs)
    axes[1,2].plot(rec_v, prec_v, label=f'PR-AUC={pr_auc:.4f}', color='red')
    axes[1,2].set_title('Precision-Recall Curve'); axes[1,2].legend()
    axes[1,2].set_xlabel('Recall'); axes[1,2].set_ylabel('Precision')
except: axes[1,2].text(0.5,0.5,'N/A',ha='center')

plt.tight_layout(); plt.show()

# --- Threshold analysis chart ---
print('\n--- Threshold Analysis ---')
thresholds = np.arange(0.1, 0.9, 0.05)
f1_vuln_list, prec_list, rec_list = [], [], []
for t in thresholds:
    preds_t = [1 if p > t else 0 for p in test_probs]
    f1_vuln_list.append(f1_score(test_labels, preds_t, pos_label=1, zero_division=0))
    prec_list.append(precision_score(test_labels, preds_t, pos_label=1, zero_division=0))
    rec_list.append(recall_score(test_labels, preds_t, pos_label=1, zero_division=0))

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(thresholds, f1_vuln_list, label='Vuln F1', linewidth=2, color='red')
ax.plot(thresholds, prec_list, label='Vuln Precision', linestyle='--')
ax.plot(thresholds, rec_list, label='Vuln Recall', linestyle='--')
ax.axvline(x=best_thresh, color='green', linestyle=':', label=f'Best thresh={best_thresh:.2f}')
ax.axvline(x=0.5, color='gray', linestyle=':', alpha=0.5, label='Default=0.50')
ax.set_xlabel('Threshold'); ax.set_ylabel('Score')
ax.set_title('Threshold vs Vuln Metrics')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

# --- LUU MODEL ---
SAVE_PATH = os.path.join(SAVE_DIR, 'vuln_rgcn_imbalance_fixed.pth')
HISTORY_PATH = os.path.join(SAVE_DIR, 'training_history_v2.json')

torch.save({
    'model_state_dict': model_full.state_dict(),
    'num_node_types': num_node_types, 'num_edge_types': num_edge_types,
    'num_code_tokens': num_code_tokens, 'num_names': num_names,
    'max_code_tokens': max_code_tokens,
    'hidden_dim': HIDDEN_DIM, 'num_layers': NUM_LAYERS,
    'dropout': DROPOUT, 'code_emb_dim': CODE_EMB_DIM,
    'best_threshold': best_thresh,
    'test_acc': test_acc_opt, 'test_f1_vuln': test_f1v_opt,
    'test_roc_auc': roc_auc, 'test_pr_auc': pr_auc,
    'best_epoch': best_epoch,
}, SAVE_PATH)

with open(HISTORY_PATH, 'w') as f:
    json.dump(history, f, indent=2)

print(f'\nModel: {SAVE_PATH}')
print(f'History: {HISTORY_PATH}')
print(f'Best threshold: {best_thresh:.2f}')
```

```python
# ===========================================
# CELL 12: PREDICTION (DUNG TUNED THRESHOLD)
# ===========================================

def predict_single(model, graph_data, device, threshold=0.5):
    model.eval()
    loader = DataLoader([graph_data], batch_size=1)
    batch = next(iter(loader)).to(device)
    with torch.no_grad():
        logit = model(batch)
        prob = torch.sigmoid(logit).item()
    label = 1 if prob > threshold else 0
    confidence = prob if label == 1 else (1 - prob)
    return label, prob, confidence

print('=' * 60)
print(f'VI DU DU DOAN (threshold={best_thresh:.2f})')
print('=' * 60)

correct = 0
total = min(10, len(test_set))
for i in range(total):
    sample = test_set[i]
    true_label = int(sample.y.item())
    pred_label, prob, conf = predict_single(model_full, sample, device, threshold=best_thresh)
    status = 'DUNG' if pred_label == true_label else 'SAI'
    if pred_label == true_label: correct += 1
    print(f'Sample {i}: Nodes={sample.x.shape[0]:4d} Edges={sample.edge_index.shape[1]:5d} | '
          f'True:{"Vuln" if true_label else "Safe":4s} | '
          f'Pred:{"Vuln" if pred_label else "Safe":4s} (p={prob:.4f} c={conf:.4f}) [{status}]')
print(f'\nDung {correct}/{total}')

# --- Tu file .pt ---
print('\n--- Du doan tu file .pt ---')
sample_file = os.path.join(DATA_DIR, pt_files[0])
sg = torch.load(sample_file, weights_only=False)
if not hasattr(sg, 'code_tokens') or sg.code_tokens is None:
    sg.code_tokens = torch.zeros(sg.x.shape[0], max_code_tokens, dtype=torch.long)
if sg.x.dim() == 1: sg.x = sg.x.view(-1, 1)
if sg.x.shape[1] == 1:
    sg.x = torch.cat([sg.x, torch.zeros(sg.x.shape[0], 3, dtype=torch.long)], dim=1)

pl, pr, pc = predict_single(model_full, sg, device, threshold=best_thresh)
print(f'File: {pt_files[0]}')
print(f'  Result: {"VULNERABLE" if pl else "SAFE"} (prob={pr:.4f}, confidence={pc:.4f})')
print(f'  Threshold: {best_thresh:.2f}')
```
